### Analisis de casos iniciados/Ministerio Publico Republica Argentina (2013/2023)

# EDA principal para la utlizacion de los datos correctos.


```python
import pandas as pd
import os
import glob

ruta_datos = os.path.join('..', 'data', 'raw')
archivos_csv = glob.glob(os.path.join(ruta_datos, "*.csv"))

print(f"Se encontraron {len(archivos_csv)} archivos.")

dfs = []
for archivo in archivos_csv:
    print(f"Leyendo: {os.path.basename(archivo)}")
    df_temp = pd.read_csv(archivo, low_memory=False)
    dfs.append(df_temp)

df_raw = pd.concat(dfs, ignore_index=True)

display(df_raw.head())
print("Combinación OK:", df_raw.shape)
```

```python

# Conteo de valores nulos por columna
null_counts = df_raw.isnull().sum().sort_values(ascending=False)
print(null_counts)

print("="*50)
# Porcentaje de nulos por columna
pct_null = (df_raw.isnull().mean() * 100).sort_values(ascending=False)
print(pct_null)
```

### CAPA 1 (typed)
En esta etapa tipificamos los datos pero sin imputar

```python
from pathlib import Path

DATA_DIR = Path("..") / "data"
TYPED_DIR = DATA_DIR / "typed"
TYPED_DIR.mkdir(parents=True, exist_ok=True)

DATE_COLS = ["caso_fecha_inicio", "caso_fecha_hecho", "fecha_envio"]
INT_COLS = [
    "provincia_id",
    "caso_cantidad_autores_identificados",
    "caso_cantidad_autores_no_identificados",
    "caso_cantidad_menores_involucrados_identificados",
    "circunscripcion_id",
    "unidad_id",
    "caso_id",
]

def to_typed(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Fechas
    for c in DATE_COLS:
        if c in df.columns:
            df[c] = pd.to_datetime(df[c], errors="coerce")

    # Numéricos enteros con nulos
    for c in INT_COLS:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce").astype("Int64")

    # Strings seguras para el resto de columnas categóricas clave
    for c in ["provincia_nombre","circunscripcion_descripcion","unidad_descripcion",
              "delito_codigo","delito_descripcion","delito_tentativa","delito_estadistico",
              "archivo_origen","caso_hora_hecho"]:
        if c in df.columns:
            df[c] = df[c].astype("string")

    return df

df_typed = to_typed(df_raw)

typed_path = TYPED_DIR / "casos_typed.csv"
df_typed.to_csv(typed_path, index=False)
print("Guardado:", typed_path, df_typed.shape)
```

### Capa 2 (EDA): imputar lo que corresponde + banderas

# Basado en tus nulos:

    -conteos → 0

    -categóricos → “No informado”

    -fechas → se mantienen null

    -hora → no imputar, pero sí crear flag

```python
EDA_DIR = DATA_DIR / "eda"
EDA_DIR.mkdir(parents=True, exist_ok=True)

COUNT_COLS_TO_ZERO = [
    "caso_cantidad_autores_identificados",
    "caso_cantidad_autores_no_identificados",
    "caso_cantidad_menores_involucrados_identificados",
]

CATEGORICAL_COLS_TO_LABEL = [
    "unidad_descripcion",
    "delito_codigo",
    "delito_descripcion",
    "delito_tentativa",
    "circunscripcion_descripcion",
]

def to_eda(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Flags de conocimiento de fecha/hora
    if "caso_hora_hecho" in df.columns:
        df["hora_hecho_conocida"] = df["caso_hora_hecho"].notna().astype("Int64")

    if "caso_fecha_hecho" in df.columns:
        df["fecha_hecho_conocida"] = df["caso_fecha_hecho"].notna().astype("Int64")

    # Conteos -> 0
    for c in COUNT_COLS_TO_ZERO:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0).astype("Int64")

    # Categóricos -> No informado
    for c in CATEGORICAL_COLS_TO_LABEL:
        if c in df.columns:
            df[c] = df[c].astype("string").fillna("No informado")

    # delito_tentativa: evitar asumir "NO"
    if "delito_tentativa" in df.columns:
        df["delito_tentativa"] = df["delito_tentativa"].replace({"": "No informado"})

    return df

df_eda = to_eda(df_typed)

eda_path = EDA_DIR / "casos_eda.csv"
df_eda.to_csv(eda_path, index=False)
print("Guardado:", eda_path, df_eda.shape)
```

### Capa 3 (ANALYTICS)

CSVs preparados para consumir.

```python
ANALYTICS_DIR = DATA_DIR / "analytics"
ANALYTICS_DIR.mkdir(parents=True, exist_ok=True)

def analytics_temporal(df: pd.DataFrame) -> pd.DataFrame:
    out = df[df["caso_fecha_inicio"].notna()].copy()
    out["anio_inicio"] = out["caso_fecha_inicio"].dt.year.astype("Int64")
    out["mes_inicio"] = out["caso_fecha_inicio"].dt.month.astype("Int64")
    return out

def analytics_geografico(df: pd.DataFrame) -> pd.DataFrame:
    # mínimo para geo
    out = df.dropna(subset=["provincia_id", "provincia_nombre"]).copy()
    return out

def analytics_penal(df: pd.DataFrame) -> pd.DataFrame:
    # mínimo para tipificación
    out = df.copy()
    if "delito_codigo" in out.columns:
        out = out[out["delito_codigo"].notna() & (out["delito_codigo"] != "No informado")].copy()
    return out

df_temporal = analytics_temporal(df_eda)
df_geo = analytics_geografico(df_eda)
df_penal = analytics_penal(df_eda)

p1 = ANALYTICS_DIR / "casos_temporal.csv"
p2 = ANALYTICS_DIR / "casos_geografico.csv"
p3 = ANALYTICS_DIR / "casos_penal.csv"

df_temporal.to_csv(p1, index=False)
df_geo.to_csv(p2, index=False)
df_penal.to_csv(p3, index=False)

print("Exportados:")
print(" -", p1, df_temporal.shape)
print(" -", p2, df_geo.shape)
print(" -", p3, df_penal.shape)
```
